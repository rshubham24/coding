There are 4 features: sepal length, sepal width, petal length, and petal width.

Start by creating 2 clusters. Then standardize the data and try again. Does it make a difference?

Use the Elbow rule to determine how many clusters are there.


## Import the relevant libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'iris_dataset.csv'</i>.

```python
raw_data = pd.read_csv('iris-dataset.csv')
data = raw_data.copy()
data
```

## Plot the data

For this exercise, try to cluster the iris flowers by the shape of their sepal. 

<i> Use the 'sepal_length' and 'sepal_width' variables.</i> 

```python
plt.scatter(data['sepal_length'], data['sepal_width'])
plt.xlabel('sepal_length')
plt.ylabel('sepla_width')
plt.show()
```

```python
def cluster_method(n, x):
    kmeans = KMeans(n)
    kmeans.fit(x)
    cluster = data.copy()
    cluster['Cluster'] = kmeans.fit_predict(x)
    plt.scatter(cluster['sepal_length'], cluster['sepal_width'], c = cluster['Cluster'], cmap='rainbow')
    plt.xlabel('sepal_length', fontsize = 20)
    plt.ylabel('sepla_width', fontsize = 20)
    plt.show()
```

# Clustering (unscaled data)

Separate the original data into 2 clusters.

```python
x = data.copy()
cluster_method(2, x)
```

# Standardize the variables

Import and use the <i> method </i> function from sklearn to standardize the data. 

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

# Clustering (scaled data)

```python
cluster_method(2, x_scaled)
```

## Take Advantage of the Elbow Method

### WCSS

```python
wcss = []
for i in range(1, 10):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
    
wcss
plt.plot(range(1, 10), wcss)
plt.show()
```

### The Elbow Method

```python
cluster_method(3, x_scaled)
```

```python
cluster_method(5, x_scaled)
```

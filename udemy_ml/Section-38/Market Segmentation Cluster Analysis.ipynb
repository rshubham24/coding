```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
raw_data = pd.read_csv('/home/rshubham/STUDY/udemy_ml/Section-38/S38_L265/3.12. Example.csv')
data = raw_data.copy()
data
```

# Plot the data

```python
plt.scatter(data['Satisfaction'], data['Loyalty'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

```python
x = data.copy()
```

# Clustering

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
clusters = x.copy()
clusters['Cluster'] = kmeans.fit_predict(x)
```

```python
plt.scatter(clusters['Satisfaction'], clusters['Loyalty'], c = clusters['Cluster'], cmap ='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

# Standardized the data

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

# Elbow method

```python
wcss = []
for i in range(1, 10):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
wcss
```

```python
plt.plot(range(1, 10), wcss)
plt.show()
```

# Making function for clustering

```python
def cluster_method(n):
    kmeans = KMeans(n)
    kmeans.fit(x_scaled)
    cluster = x.copy()
    cluster['Cluster'] = kmeans.fit_predict(x_scaled)
    plt.scatter(cluster['Satisfaction'], cluster['Loyalty'], c = cluster['Cluster'], cmap='rainbow')
    plt.xlabel('Satisfaction', fontsize = 20)
    plt.ylabel('Loyalty', fontsize = 20)
    plt.show()
```

# Exploring cluster analysis using the results of elbow method

```python
cluster_method(2)
```

```python
cluster_method(3)
```

```python
cluster_method(4)
```

```python
cluster_method(5)
```

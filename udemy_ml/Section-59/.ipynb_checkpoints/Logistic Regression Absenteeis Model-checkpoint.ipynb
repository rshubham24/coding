# Create the Target

# Standardized the input

```python
#from sklearn.preprocessing import StandardScaler
```

```python
#scaler = StandardScaler()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

# create the Custom Scaler class

class CustomScaler(BaseEstimator,TransformerMixin): 
    
    def __init__(self,columns,copy=True,with_mean=True,with_std=True):
        self.scaler = StandardScaler(copy,with_mean,with_std)
        self.columns = columns
        self.mean_ = None
        self.var_ = None
        
    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns], y)
        self.mean_ = np.mean(X[self.columns])
        self.var_ = np.var(X[self.columns])
        return self
    
    def transform(self, X, y=None, copy=None):
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        X_not_scaled = X.loc[:,~X.columns.isin(self.columns)]
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
```

```python
columns_to_omit = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Education']
```

```python
columns_to_scale = [x for x in input_data.columns.values if x not in columns_to_omit]
```

```python
scaler = CustomScaler(columns_to_scale)
```

```python
scaler.fit(input_data)
```

```python
scaled_inputs = scaler.transform(input_data)
```

# Split the data into train & test and shuffle

```python
from sklearn.model_selection import train_test_split 
```

```python
 x_train, x_test, y_train, y_test = train_test_split(scaled_inputs, targets, test_size = 0.2, random_state = 20)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

# Logistic Regression with sklearn

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train, y_train)
```

```python
reg.score(x_train, y_train)
```

## Manualy check the accuracy

```python
model_outputs = reg.predict(x_train)
```

```python
np.sum(model_outputs == y_train)
```

## Finding the intercepts and coefficient

```python
feature_name = input_data.columns.values
```

```python
summary_table = pd.DataFrame(columns = ['Feature Name'], data = feature_name)
summary_table['Coefficient'] = np.transpose(reg.coef_)
summary_table
```

```python
summary_table.index = summary_table.index + 1
summary_table.loc[0] = ['Intercept', reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

# Interpreting the coefficient

```python
summary_table['Odds_ratio'] = np.exp(summary_table['Coefficient'])
summary_table
```

```python
summary_table.sort_values('Odds_ratio', ascending = False) # If odds_ratio is nearly 1 then that feature is not important
```

# Testing the model

```python
reg.score(x_test, y_test)
```

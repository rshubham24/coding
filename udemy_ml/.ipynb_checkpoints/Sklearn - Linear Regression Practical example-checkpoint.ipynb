# Relaxing the Assumptions

```python
log_price = np.log(data_cleaned['Price'])
```

```python
data_cleaned['log_price'] = log_price
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(15, 3))
ax1.scatter(data_cleaned['Year'], data_cleaned['log_price'])
ax1.set_title('log_price and Year')
ax2.scatter(data_cleaned['EngineV'], data_cleaned['log_price'])
ax2.set_title('log_price and EngineV')
ax3.scatter(data_cleaned['Mileage'], data_cleaned['log_price'])
ax3.set_title('log_price and Mileage')
plt.show()
```

```python
data_cleaned = data_cleaned.drop(['Price'], axis = 1)
```

# Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage', 'Year', 'EngineV']]
vif = pd.DataFrame()
vif['features'] = variables.columns
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif
```

```python
data_no_multicollinearity = data_cleaned.drop(['Year'], axis=1)
```

```python
data_no_multicollinearity.describe(include = 'all')
```

# Create Dummy Variables

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first = True)
```

```python
data_with_dummies
```

# Rearrange a bit

```python
data_with_dummies.columns.values
```

```python
cols = ['log_price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed
```

# Linear Regression Model

## Declare the inputs and the targets

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'], axis=1)
```

## Scale the Data

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
Scaler.fit(inputs)
```

```python
input_scaled = Scaler.transform(inputs)
```

## Train and Test Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_scaled, targets, test_size = 0.2, random_state = 365)
```

```python
x_train
```

## Create the Regression

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat)
plt.xlabel('Targets(y_train)', size=18)
plt.ylabel('Predictions(y_hat)', size=18)
plt.xlim(6, 13)
plt.ylim(6, 13)
plt.show()
```

```python
sb.distplot(y_train - y_hat)
plt.title('Residual PDF', size=18)
plt.show()
```

```python
reg.score(x_train, y_train)
```

```python
reg.intercept_
```

```python
reg_summary = pd.DataFrame()
reg_summary['Features'] = inputs.columns
reg_summary['Weights'] = reg.coef_
reg_summary
```

# Testing

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test, y_hat_test, alpha = 0.4)
plt.xlabel('Targets(y_test)', size=18)
plt.ylabel('Predictions(y_test)', size=18)
plt.xlim(6, 13)
plt.ylim(6, 13)
plt.show()
```

```python
sb.distplot(y_test - y_hat_test)
plt.title('Residual PDF', size=18)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns = ['Prediction'])
```

```python
y_test = y_test.reset_index(drop=True)
```

```python
df_pf['Target'] = np.exp(y_test)
```

```python
df_pf.head()
```

```python
df_pf['Residual'] = df_pf['Target'] - df_pf['Prediction']
df_pf['Differences%'] = np.absolute(df_pf['Residual'])/df_pf['Target']*100
```

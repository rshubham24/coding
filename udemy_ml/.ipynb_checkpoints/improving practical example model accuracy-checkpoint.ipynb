```python
data_cleaned_no_mc = data_cleaned.drop(['Year', 'Brand', 'EngineV', 'Registration'], axis=1)
data_cleaned_no_mc.describe(include = 'all')
```

```python
data_with_dummies = pd.get_dummies(data_cleaned_no_mc, drop_first = True)
data_with_dummies.describe()
```

```python
culs = data_with_dummies.drop(['Price', 'Model_323', 'Model_4Runner','Model_524','Model_545','Model_CLA-Class', 'Body_sedan'], axis = 1)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variable = culs
vif = pd.DataFrame()
vif['features'] = variable.columns
vif['VIF'] = [variance_inflation_factor(variable.values, i) for i in range(variable.shape[1])]
vif.sort_values(['VIF'], axis=0, ascending=False, inplace=True)
pd.options.display.max_rows = 999
vif
```

# Linear Regression MOdel

```python
targets = data_with_dummies['Price']
inputs = data_with_dummies.drop(['Price', 'Model_323', 'Model_4Runner','Model_524','Model_545','Model_CLA-Class', 'Body_sedan'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs, targets, test_size = 0.2, random_state = 365)
```

```python
x_test.shape
x_test
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
x_train.shape
```

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
reg.score(x_train, y_train)
```

```python
coef = pd.DataFrame()
coef['Features'] = x_train.columns.values
coef['Weights'] = reg.coef_
coef.sort_values(['Weights'], axis = 0, ascending = False, inplace = True)
coef
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train, y_hat, alpha=0.4)
plt.xlabel('Targets(y_train)', size=18)
plt.ylabel('Predictions(y_hat)', size=18)
plt.show()
```

```python
sb.distplot(y_train - y_hat)
plt.title('Residual PDF', size=18)
plt.show()
```

# Testing

```python
y_hat_test = reg.predict(x_test)
y_hat = pd.DataFrame()
```

```python
y_hat['y_hat_test'] = y_hat_test
y_test = y_test.reset_index(drop = True)
y_hat['y_test'] = y_test
sb.distplot(y_hat['y_hat_test'])
```

```python
q = y_hat['y_hat_test'].quantile(0.99)
y_hat = y_hat[y_hat['y_hat_test'] < q]
sb.distplot(y_hat['y_hat_test'])
```

```python
q = q = y_hat['y_hat_test'].quantile(0.01)
y_hat = y_hat[y_hat['y_hat_test'] > q]
sb.distplot(y_hat['y_hat_test'])
```

```python
plt.scatter(y_hat['y_test'], y_hat['y_hat_test'], alpha = 0.4)
plt.xlabel('Targets(y_test)', size=18)
plt.ylabel('Predictions(y_test)', size=18)
plt.show()
```

```python
df_pf = pd.DataFrame()
df_pf['Prediction'] = y_hat['y_hat_test'] 
df_pf['Target'] = y_hat['y_test']
df_pf['Residual'] = df_pf['Target'] - df_pf['Prediction']
df_pf['Differences%'] = np.absolute(df_pf['Residual'])/df_pf['Target']*100
df_pf.describe()
```

# Regression with price with brand

```python
reg = LinearRegression()
reg.fit(x_train_price_brand, y_train_price_brand)
```

```python
y_hat_price_brand = reg.predict(x_train_price_brand)
```

```python
plt.scatter(y_train_price_brand, y_hat_price_brand)
plt.xlabel('Targets(y_train_price_brand)', size=18)
plt.ylabel('Predictions(y_hat_price_brand)', size=18)
plt.show()
```

```python
sb.distplot(y_train_price_brand - y_hat_price_brand)
plt.title('Residual PDF', size=18)
plt.show()
```

```python
reg.score(x_train_price_brand, y_train_price_brand)
```

```python
reg.intercept_
```

```python
reg_summary = pd.DataFrame()
reg_summary['Features'] = input_price_brand.columns
reg_summary['Weights'] = reg.coef_
reg_summary
```

```python
df_pf_price_brand = pd.DataFrame(y_hat_test_price_brand, columns = ['Prediction'])
```

```python
y_test_price_brand = y_test_price_brand.reset_index(drop=True)
```

```python
df_pf_price_brand['Target'] = y_test_price_brand
```

```python
df_pf_price_brand['Residual'] = df_pf_price_brand['Target'] - df_pf_price_brand['Prediction']
df_pf_price_brand['Differences%'] = np.absolute(df_pf_price_brand['Residual'])/df_pf_price_brand['Target']*100
```

```python
df_pf_price_brand.describe().round(2)
```

```python
pd.options.display.max_rows = 999
pd.set_option('display.float_format', lambda x: '%2f' %x)
df_pf_price_brand.sort_values(by=['Differences%'])
```

# Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# Generate input

```python
observations = 1000
xs = np.random.uniform(-10,10,(observations, 1))
zs = np.random.uniform(-10,10,(observations, 1))
```

```python
inputs = np.column_stack((xs,zs))
print(inputs.shape)
inputs
```

# Generate target

```python
noise = np.random.uniform(-1,1,(observations, 1))
targets = 2*xs - 3*zs + 5 + noise
targets
```

# Plot the training data

```python
targets = targets.reshape(observations,)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs, zs, targets)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('Targets')
ax.view_init(azim=100)
plt.show()
targets = targets.reshape(observations,1)
```

# Initialize variable

```python
init_range = 0.1
weight = np.random.uniform(-init_range, init_range,(2,1))
buise = np.random.uniform(-init_range, init_range, (1))
print(weight)
print(buise)
```

```python
learning_rate = 0.02
```

# Train the model

```python
for i in range (1000):
    outputs = np.dot(inputs, weight) + buise
    deltas = outputs - targets
    loss = np.sum(deltas ** 2)/2/observations
    print(loss)
    deltas_scaled = deltas/observations
    weight = weight - learning_rate*np.dot(inputs.T, deltas_scaled)
    buise = buise - learning_rate*np.sum(deltas_scaled)
```

```python
print(weight, buise)
```

# Plot last output vs target

```python
plt.plot(outputs,targets)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
